```python
#Linear Regression for general functions
import numpy as np
import matplotlib.pyplot as plt
import math
pi=math.pi


x0=np.linspace(-1,1,100)

y0=np.sin(2*pi*5*x0)+0.0*np.random.randn(x0.size)
plt.plot(x0,y0)
```

```python
##I've try plenty of times, when Num > 38 will cause overfit
Num=38
basis=np.arange(Num)
x0=x0[:,np.newaxis]
basis=basis[np.newaxis,:]
A=x0**basis
print(A.shape)
y0=y0[:,np.newaxis]
w=np.linalg.inv(A.T@A)@(A.T@y0)
yhat=A@w
plt.plot(x0,yhat,x0,y0)
```

```python
x0=np.linspace(0.1,1,100)
y0=np.log(x0)+0.1*np.random.randn(x0.size)
plt.plot(x0,y0)
```

```python
##I guess orignal func is a log plus noise
Num=20
basis=np.arange(Num)
x0=x0[:,np.newaxis]
basis=basis[np.newaxis,:]
A=x0**basis
print(A.shape)
y0=y0[:,np.newaxis]
w=np.linalg.inv(A.T@A)@(A.T@y0)
yhat=A@w
plt.plot(x0,yhat,x0,y0)
```
