```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def plotClass(X,y,p):
    plt.figure()
    for i in range(y.shape[1]):
        if y[0,i]==0:
            plt.plot(X[0,i],X[1,i],'r'+p)
        else:
            plt.plot(X[0,i],X[1,i],'b'+p)
            
    plt.show()
```

```python
eps=10e-6
def getZ(w,x,b):
  return w@x + b

def sigmoid(z):
  return (1/(1+np.exp(-z)))

def loss(y,yhat):
  return np.sum(-y*np.log(yhat+eps)-(1-y)*np.log(1-yhat+eps), 1)
```

```python
#Q1
num_data=1000 # data points per class

X=np.random.uniform(-1,1,[2,num_data])
y=X[0,:]**2+X[1,:]**2-0.5>0
y=(y[None,:]).astype(np.int8)
plotClass(X,y,'o')
print(X.shape)
print(y.shape)
```

```python
w = np.random.random((1,9))
b = 0
ls = []
lr = 0.001
num = 3
bs = []
for i in range(num):
  for j in range(num):
    bss = (X[0,:]**i)*((X[1,:])**j)
    bs.append(bss)

basis = np.asarray(bs)

for i in range(10000):
  z=getZ(w,basis,b)
  yhat=sigmoid(z)
  l=loss(y,yhat)
  ls.append(l)
  dz=yhat-y
  dw=dz@basis.T
  db=np.sum(dz,1)
  w=w-lr*dw
  b=b-lr*db

yhat.shape
plt.plot(ls)
plotClass(X,yhat > 0.5,'.')
```
