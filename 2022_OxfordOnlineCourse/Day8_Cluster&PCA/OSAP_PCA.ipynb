```python
import numpy as np
import matplotlib.pyplot as plt


!wget https://www.dropbox.com/s/eh1kps73fp88m67/faces.dat
```

```python
faces=np.genfromtxt('faces.dat')
print(faces.shape)
```

```python
def printface(face):
  plt.imshow(face.reshape(64,64).T,cmap='gray')
```

```python
printface(faces[0])
```

```python
#perfom PCA for eigenfaces, show the reconstruction of faces project on 10,100,200 principle components
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=faces.shape[0])
pca.fit(faces)

comp=pca.components_
print(comp.shape)

for i in range(10):
  plt.figure()
  printface(comp[i])
```

```python
for num in ([10,100,200]):
  plt.figure()
  comps=range(0,num)
  print(comps)
  img=faces[0]
  imgP=np.zeros(img.shape)
  for n in comps:
    imgP+=np.dot(img,comp[n,:])*comp[n]
  printface(imgP)
```
