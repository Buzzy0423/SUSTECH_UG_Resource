```python
import numpy as np
import random
import CARP_solver
import matplotlib.pyplot as plt
import os
```

```python
folder_path = 'test/CARP_samples'
file_names = os.listdir(folder_path)
result_score = {}
result_r = {}
psize = [5, 10, 20, 50, 100, 200]
for i, f in enumerate(file_names):
    result_score[f], result_r[f] = [], []
    for size in psize:
        args = CARP_solver.preprocess(os.path.join(folder_path, f), 10, 10, size)
        s, g, t  = [], [], []
        for j in range(3):
            solution, score, gen, evolution, time, avg = CARP_solver.genetic(*args)
            s.append(score)
            g.append(gen)
            t.append(time)
        result_score[f].append(np.average(s))
        result_r[f].append(np.average(g)/np.average(t))
```

```python
for i in result_score.keys():
    plt.plot(psize,result_score[i])
    plt.title(i[2:-4])
    plt.xlabel('population size')
    plt.ylabel('best solution cost')
    plt.savefig(f'image/cost/{i[2:-4]}.png', dpi=500)
    plt.show()
```

```python
for i in result_r.keys():
    plt.plot(psize, result_r[i])
    plt.xlabel('population size')
    plt.ylabel('R')
    plt.title(i[2:-4])
    plt.savefig(f'image/g/{i[2:-4]}.png', dpi=500)
    plt.show()
```
